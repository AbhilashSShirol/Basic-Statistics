Method 2

```python
# Given information
mu = 270  # Claimed average lifespan
n = 18    # Sample size
x_bar = 260  # Sample mean
s = 90    # Sample standard deviation

# Calculating the t-score
t_score = (x_bar - mu) / (s / (n ** 0.5))

# Calculating the probability using the t-distribution
df = n - 1
probability = stats.t.cdf(t_score, df)

probability
```
